### Task 4: Set Application Follow Up Priority based on Location and Work Type. 

```python
def application_follow_up_priority(row):
    if "Data Analyst" in row["job"]:
        if row["location"] == "Bengaluru":
            return 1
        elif row["location"] == "Delhi" and row["work_type"] == "Remote":
            return 2
        else: 
            return 0
    else:
        return 0

df["follow_up_priority"] = df.apply(application_follow_up_priority, axis=1)
```

```python
df["follow_up_priority"].unique()
```

```python
follow_up_df = df.query('(follow_up_priority == 1) or (follow_up_priority == 2)')
```

```python
follow_up_df[["job", "company_name", "location", "work_type", "follow_up_priority"]]
```
