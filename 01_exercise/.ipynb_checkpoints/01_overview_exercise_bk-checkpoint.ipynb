## TASK 3: Merge Data.

```python
!ls ../01_data
```

```python
df1 = pd.read_csv("../01_data/commodity_futures_2000_to_2019.csv")
df2 = pd.read_csv("../01_data/commodity_futures_from_2020.csv")
```
